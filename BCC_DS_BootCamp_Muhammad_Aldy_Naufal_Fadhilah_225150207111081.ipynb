```python
from sklearn.model_selection import train_test_split

# memisahkan data menjadi data train dan data test
train, test = train_test_split(data, test_size=0.2, random_state=5000)

print("Banyak data: ", len(data))
print("Banyak data train: ", len(train))
print("Banyak data test: ", len(test))
```

```python
data.columns
```

```python
label_regresi = 'median_house_value'
label_klasifikasi = "housing_median_age_category"
```

```python
train_data = train.drop([label_regresi, label_klasifikasi], axis=1)

train_label_regresi = train[label_regresi]
train_label_klasifikasi = train[label_klasifikasi]
```

```python
test_data = test.drop([label_regresi,label_klasifikasi], axis=1)

test_label_regresi = test[label_regresi]
test_label_klasifikasi = test[label_klasifikasi]
```

```python
train_data.head()
```

**Scaling menggunakan Standard Scaling**



```python
data_numerik = train_data.select_dtypes(include=np.number).columns.tolist()
data_numerik
```

```python
from sklearn.preprocessing import StandardScaler

scelar = StandardScaler()

train_ss = train_data.copy()

train_ss[data_numerik]=scelar.fit_transform(train[data_numerik])
train_ss.head()
```

```python
train_ss.describe()
```

```python
from sklearn.preprocessing import StandardScaler

scelar = StandardScaler()

test_ss = test_data.copy()

test_ss[data_numerik]=scelar.fit_transform(test[data_numerik])
test_ss.head()
```

```python
test_ss.describe()
```

```python
data_object = train_ss.select_dtypes(include = object).columns.tolist()
data_object
```

Mengubah object ocean_proximity menjadi numery

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder()

train_ohe = ohe.fit_transform(train_ss[data_object]).toarray()

train_ohe = pd.DataFrame(train_ohe, columns=ohe.get_feature_names_out())

train_ss = pd.concat([train_ss.reset_index(drop=True), train_ohe], axis = 1)

train_ss = train_ss.drop(data_object, axis = 1)

train_ss.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder()

test_ohe = ohe.fit_transform(test_ss[data_object]).toarray()

test_ohe = pd.DataFrame(test_ohe, columns=ohe.get_feature_names_out())

test_ss = pd.concat([test_ss.reset_index(drop=True), test_ohe], axis=1)

test_ss = test_ss.drop(data_object, axis=1)

test_ss.head()
```

Regresi Predict of Median_house_values

Visualization of Data

```python
filter_still_new = data['housing_median_age_category'] == 'Still New'
filter_old = data['housing_median_age_category'] == 'Old'

filtered_data_still_new = data.loc[filter_still_new,['housing_median_age_category','median_house_value']]
filtered_data_old = data.loc[filter_old, ['housing_median_age_category', 'median_house_value']]

x_new = filtered_data_still_new['housing_median_age_category'].unique()
y_new = filtered_data_still_new['median_house_value'].sum()
plt.bar(x_new , y_new)

x_old = filtered_data_old['housing_median_age_category'].unique()
y_old = filtered_data_old['median_house_value'].sum()
plt.bar (x_old, y_old)

plt.xlabel('Housing Median Age Category')
plt.ylabel('Houe Value')
plt.title("Disstribusi Harga berdasarkan umur")
plt.show()
```

Visualisasi diatas merupakan perbandingan jumlah median house value berdasarkan housing median age category

```python
x = data['ocean_proximity'].unique()
y = data['ocean_proximity'].value_counts()

plt.bar(x, y)
plt.xlabel("Lokassi rumah terhadap laut")
plt.ylabel("Jumlah")
plt.title("Distribusi ocean proximity")
plt.show()
```
